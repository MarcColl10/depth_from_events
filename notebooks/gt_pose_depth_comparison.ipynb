```python
# 20 epochs
run_ids = {
    "w/o pose gt": "vhh5xu3p",
    "w/ pose gt": "ljobzfd0",
    "w/ rotation gt": "zo0h6xvi",
    "w/ translation gt": "udn5wtjo",
}
```

```python
fps = 100
depth_maps = dict()
event_clip = ImageSequenceClip(f"../logs/images/validate_{list(run_ids.values())[0]}_latest/validate/events", fps=fps)
event_clip = CompositeVideoClip([
    event_clip,
    TextClip("Event", fontsize=10, color='white').set_position(('center', 'top')).set_duration(event_clip.duration),
])
for key, run_id in run_ids.items():
    depth_clip = ImageSequenceClip(f"../logs/images/validate_{run_id}_latest/validate/disparity", fps=fps)
    depth_maps[key] = CompositeVideoClip([
        depth_clip,
        TextClip(key, fontsize=10, color='white').set_position(('center', 'top')).set_duration(depth_clip.duration),
    ])


vid_array = clips_array([[event_clip, event_clip],
                         [depth_maps["w/o pose gt"], depth_maps["w/ pose gt"]],
                         [depth_maps["w/ rotation gt"], depth_maps["w/ translation gt"]]])
vid_array.write_videofile("../logs/uzh_fpv_depth_maps_gt_pose_comparison.mp4")
```
