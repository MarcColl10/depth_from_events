```python
import matplotlib.cm as cm
from matplotlib.colors import hsv_to_rgb, Normalize
import matplotlib.pyplot as plt
import imageio
import numpy as np
```

```python
from collections import OrderedDict
indices = [2473, 3233, 4972, 8696]
sequences = OrderedDict({
    "Image": "network0/validate/color_gt",
    "GT": "network0/validate/disparity_gt",
    "Step 0": "network0/validate/disparity",
    "Step 3K (OL)": "network3000/validate/disparity",
    "Step 6K (OL)": "network6000/validate/disparity",
    "Step 9K (OL)": "network9000/validate/disparity",
    "Step 12K (OL)": "network12000/validate/disparity",
    "Step 12K (TFS)": "fromscratch12000/validate/disparity",
})
```

```python
fig, axs = plt.subplots(len(sequences), 4, figsize=(
    15, 22), constrained_layout=True)
for i, (name, folder) in enumerate(sequences.items()):
    for j, file_idx in enumerate(indices):
        ax = axs[i, j]
        if j == 0:
            ax.set_ylabel(name, fontsize=20, labelpad=15)
        ax.tick_params(
            axis='both',          # changes apply to the x-axis
            which='both',      # both major and minor ticks are affected
            left=False,      # ticks along the bottom edge are off
            top=False,
            bottom=False,  # ticks along the top edge are off
            labelbottom=False,
            labelleft=False)

        # image
        ax.imshow(plt.imread(
            f"finetune_logs/images/{folder}/{str(file_idx).zfill(5)}.png"))

fig.savefig('finetune_qual.pdf')
```

```python
steps = [0, 300, 600, 900, 1200]
RSAT = np.array([0.99, 0.95, 0.95, 0.96, 0.95])
MAE = np.array([1.47, 1.34, 1.33, 1.31, 1.31])

fig, ax2 = plt.subplots(figsize=(8, 3.5), constrained_layout=True)
ax1 = ax2.twinx()  # instantiate a second axes that shares the same x-axis
ax1.set_xticks(steps)
ax2.set_ylim(1.3, 1.5)
ax2.yaxis.set_major_locator(plt.MaxNLocator(nbins=5))
color = "C0"
ax1.set_xlabel('Learning Steps')
# ax1.set_ylabel('RSAT', color=color)
ax1.plot(steps, RSAT, color="C0", marker='o')
ax1.tick_params(axis='y', labelcolor=color)


ax1.set_ylim(0.9, 1.0)
color = 'C1'
# ax2.set_ylabel('MAE', color=color)  # we already handled the x-label with ax1
ax2.plot(steps, MAE, color="C1", marker='o')
ax2.set_xlabel('Learning Steps', fontsize=14)
ax1.legend(['RSAT'], loc='upper right', fontsize=14)
ax2.legend(['MAE'], loc='lower left', fontsize=14)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
ax1.tick_params(axis='both', which='major', labelsize=14)
ax2.tick_params(axis='both', which='major', labelsize=14)
# ax2.legend(['MAE'], loc='upper left')

plt.savefig('finetune_metrics.pdf')
```
